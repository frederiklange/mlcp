## 1.4.1 Answer

First, I stretch the static features in order to merge them with the UCDP. To do so, I start by grouping the data by year and priogrid-gid. This results in data where death counts is counted yearly on a grid cell level. Then I rename the priogrid_gid so it matches the names in the PRIO data - gid.

```python
# Grouping the data by priogrid_gid and year
ucdp_gid = ucdp.groupby(by=['priogrid_gid','year']).sum().reset_index()

# Renaming the priogrid_gid to gid 
ucdp_gid.rename(columns={'priogrid_gid':'gid'}, inplace=True) # Inplace is equal to True in order to keep the changes to the dataframe

# Creating a list with all the unique years in the data - sorted.
years = sorted(ucdp_gid['year'].unique())
```

I add a feature with 'years' to the grid_static dataframe. This allows me to create a loop where I add copies of all observations of the static data to the original grid_static data with a year feature attached. The first copy will have 1989 attached, however, each new copy will have a new year. The data will be similar for all years, but this is not a problem since the data is static. For example, mountains do not change that much over a time span of 30 years.

```python
# Copying the dataframe to get the initial year 
grid_static_yearly = grid_static.copy() 

# Creating feature called year with 1989 as a starting point
grid_static_yearly['year'] = years[0]

# Copying the data frame in order to have a new dataframe to concatenate 
grid_static_one_year = grid_static_yearly.copy() 

# Creating a loop over the years list created earlier. In each loop I add a new year to the grid_static_yearly dataframe with the static data
for i in years[1:]: 
    df_temp = grid_static_one_year.copy() # Creating a temporary dataframe
    df_temp['year'] = i # Setting the value of the temporary dataframe to year i
    grid_static_yearly = pd.concat([grid_static_yearly, df_temp]) # Concatenating the new frame to the original data

# Resetting the index so it does not just repeat itself for each year
grid_static_yearly.reset_index(inplace = True) 
```

To make sure, that the UCDP and the geopandas dataframe match on the temporal dimension. I check the unique years of both dataframes with a logical test.

```python
# Checking if the two columns have the same unique number of years
grid_static_yearly['year'].unique() == ucdp_gid.sort_values(by = ['year'])['year'].unique() # I sort the values of the ucdp_gid by year in order to make the compared arrays testable with a logical operator 
```

Now that it is clear, that both dataframes contains the same years. I finally merge the yearly, static data set with the UCDP data. I perform a left merge on 'gid' and 'year'. This means that if there is conflict regarding the land mass defined by the PRIO static data, I will just discard it.

```python
# Merging the static data and ucdp data
grid_ucdp =  pd.merge(grid_static_yearly, ucdp_gid, how = 'left', on = ['gid', 'year'])

# Checking the shape of the dataframe
grid_ucdp.shape 
```

Now, to complete our data set we only need to add the dynamic features. But as you know, the PRIO dynamic variables are riddled with missing values. Lets get rid of these first using linear interpolation and extrapolation (fill backward and fill forward). Again, this can be done in many ways, one which I have already provided you. 

To save computation time and keep our effort concise, you only need to concentrate this interpolation/extrapolation effort on a subset of features (features_of_interest):
- ['nlights_mean', 'pop_gpw_sum', 'agri_ih', 'barren_ih', 'urban_ih', 'petroleum_y','excluded']   

Test whether you succeeded. Do this by plotting the trend of the (interpolated) "nlights_mean" features throughout the years. Do this for 8 more or less random gids in the same plot (for a more interesting plot make sure you draw gids with at least some non-zero values on the (interpolated) "nlights_mean" feature).

 ## 1.6.1 Answer

First, I import scipy which I use to interpolate/extrapolate, and time in order to calculate the run time of the interpolation/extrapolation.

```python
# Importing scipy and time
import scipy.interpolate
import time
```

Second, I create the interpolation/extrapolation function that I will use to interpolate/extrapolate the dynamic data

```python
# Creating interpolation/extrapolation function
def interpolation_function(features_to_interpolate, years, all_gids, dataset):
    """This function interpolates/extrapolates dynamic data by taking the features that needs to be interpolated/extrapolated,
    the years that need to be interpolated/extrapolated, the gids that are in the dataset and the dataset that is used.
    The function returns the dataset with interpolated values"""
    # Creating variables with number of gids and number of features
    n_gids = len(all_gids)
    n_features = len(features_to_interpolate)

    # Outer loop over the features 
    for j, feature in enumerate(features_to_interpolate):
        
        # Inner loop over the alle the gids
        for i, gid in enumerate(all_gids):

            # Printing status of the functionr
            print(f'processing feature ({feature}) {j+1}/{n_features} : gid ({gid}) {i+1}/{n_gids}........', end='\r') 

            # Creating a small slice of our dataset with only the actual observations and the corrospoding years
            observations = dataset[dataset['gid']==gid][[feature,'year']].dropna()

            # Creating X (years) and y (feature)
            X = observations.iloc[:,1]
            y = observations.iloc[:,0]

            # Creating try/except statement. The function first try to interpolate. If an error occurs it moves to the except part.
            try:
                interpolationFunction = scipy.interpolate.interp1d(X, y, bounds_error = False)
                y = interpolationFunction(years) 

            except:
                X = years
                y = np.zeros(len(years)) # If we get an interpolation error we pass the observation with a value of zero.
            
            # Creating temporary dataframe
            temp_df =  pd.DataFrame([y, years]).T.fillna(method='ffill').fillna(method='bfill')# ffill fills forward from lat obs. Then bfill fill backwards from first obs

            # Creating a new name for our interpolated feature.
            new_feature_name = 'interp_' + feature

            # Inserting the new feature into the old dataset
            observations_index = dataset[dataset['gid'] == gid].index
            dataset.loc[observations_index,new_feature_name] = temp_df[0].values
    
    # Returning the interpolated dataset
    return dataset
```

Third, I create the different data elements needed for the function created above:
- list for the subset of features that is outlined above

First, I create the training, validation and test data sets. The test data set will contain all values from 2014. I create the data set by creating masks that contain all the indices for each of the data sets. Afterwards, I use the masks to filter the data and create the different data sets. 

```python
# Creating masks that filters the indexes of test, val and train
test_idx = trim_df[trim_df['year'] == 2014].index # get the indices from all observation in year 2014
val_idx = trim_df[trim_df['year'] == 2013].index # get the indices from all observation in year 2013
not_train_idx = np.append(test_idx,val_idx) # get indices from all observations that are not test or val

# Creating the test, val and train set from the masks
test_set = trim_df.loc[test_idx] # use the 2014 indecies to make a test set
val_set = trim_df.loc[val_idx] # use the 2013 indecies to make a validation set
train_set = trim_df.drop(index = not_train_idx) # Create training set containing non of the indecies from 2013 or 2014.
```

We create the training, validation and test sets in order to better evaluate our models performance and improve the performance as well. The basic idea is to seperate the data used to calculate/fit the model from the data used to critisize the performance of the fitted model (Friedman et al., 2001: 222; Colaresi & Mahmood, 2017).

In our case, the training data will be used to train our model before evaluating the model on the validation data. This allows us to get a feeling for our models out-of-sample performance. For example, we use the evaluation of the model performance on the validation data to fine tune our hyperparameters before finally evaluating our model on the test set to see how our model actually performs. 

If we did not evaluate our model on data that are different from the data we use to fit the model, we will only know the models in-sample prediction. This prediction may be biased due to e.g. overfitting to random patterns of noise in the sample data (Colaresi & Mahmood, 2017). Therefore, our model will be very prone to perform badly on out-of-sample data, put differently, it generalizes badly. However, it is hard to evaluate if our model is overfitted when we only observe its performance on in-sample data. The out-of-sample predictions on the validation data allows us to somewhat overcome this problem, and thereby readjust parameters to improve the out-of-sample performance of our model.

In our case, we need to be careful of the fact that our data is time series based. Time series data have a lot of autocorrelation, and therefore we can not just randomnly split our data into test, validation and training data. We are interested in predicting the future, so the out-of-sample predictions has to be predictions of future conflicts. If we do a random split of our data into training, validation and test data, we risk that our model is trained on data from the future which is not possible if we try to predict conflicts that have not happened yet. Therefore, we need to split our data based on time - just as we did above. In sum, the out-of-sample predictions and potential retuning of the hyperparameters of our model need to be done on data seperated in time. 

It should be noted however, that the fine tuning of hyper parameters with k-fold cross validiation and grid-search can possibly benefit from mixing time within the training set. In that case, we are not interested in the absolute performance of the model. Instead, we want to now the hyperparameters that provide the best results. It can therefore be argued, that as long as the model is tested on time seperated data out of sample, the process of tuning hyperparameters may or may not benefit from time being mixed - this is an emperical question.

## 1.13.1 Answer

I create the features using three functions. The first two functions relates to the concept of the conflict trap where conflicts are believed to have a degree of intertia. To reflect this intertia, however, the feature needs to deterioate over time. The defined functions create an exponential detoriation index with detoriation over 5 and 10 years respectively. The last function incorporates these two functions and creates the rest of the different features. I briefly comment and reflect on each of the features created.

```python
# Creating the function with a 5 year decay
def decay5(data, window=5):
    """Creates a 5 year exponential detoriation index by taking a subset of data and the number of years over which it should decay.
    The default period is 5 years"""
    alpha = 2 /(window + 1.0)
    alpha_rev = 1-alpha
    n = data.shape[0]

    pows = alpha_rev**(np.arange(n+1))

    scale_arr = 1/pows[:-1]
    offset = data.iloc[0]*pows[1:]
    pw0 = alpha*alpha_rev**(n-1)

    mult = data*pw0*scale_arr
    cumsums = mult.cumsum()
    out = offset + cumsums*scale_arr[::-1]
    return out
```

```python
# Creating the function with a 10 year decay
def decay10(data, window=10):
    """Creates a 10 year exponential detoriation index by taking a subset of data and the number of years over which it should decay.
    The default period is 10 years"""
    alpha = 2 /(window + 1.0)
    alpha_rev = 1-alpha
    n = data.shape[0]

    pows = alpha_rev**(np.arange(n+1))

    scale_arr = 1/pows[:-1]
    pw0 = alpha * alpha_rev**(n-1)

    mult = data*pw0*scale_arr
    cumsums = mult.cumsum()
    out = cumsums * scale_arr[::-1]
    return out
```

```python
def featureEng(test_time = False):
    """The function creates multiple features from the existing data in our training, validation and test data."""
    pkl_file = open('train_set.pkl', 'rb')
    train_set = pickle.load(pkl_file)
    pkl_file.close()
    
    if test_time == True:
        
        pkl_file = open('val_set.pkl', 'rb')
        val_set = pickle.load(pkl_file)
        pkl_file.close()
    
        pkl_file = open('test_set.pkl', 'rb')
        test_set = pickle.load(pkl_file)
        pkl_file.close()
    
        df = pd.concat([train_set, val_set, test_set])

    
    elif test_time == False:
    
        
        pkl_file = open('val_set.pkl', 'rb')
        val_set = pickle.load(pkl_file)
        pkl_file.close()
    
        df = pd.concat([train_set, val_set])
    
    else:
        
        print("unrecognized input for test_time...")
        
    df.sort_values('year', inplace = True)

    # ----------------------------------------------------------------------------------------------
    # Creating features that measures past conflict in different ways in a given cell or country.
    # The features are created by taking the cumulative sum of 'best', 'log_best' and 'binary_best' over time.
    df['past_fatalities'] = df.sort_values('year').groupby(['gid'])['best'].cumsum()
    df['past_magnitude'] = df.sort_values('year').groupby(['gid'])['log_best'].cumsum()
    df['past_events'] = df.sort_values('year').groupby(['gid'])['binary_best'].cumsum()

    df['past_fatalities_country'] = df.sort_values('year').groupby(['gwno'])['best'].cumsum()
    df['past_magnitude_country'] = df.sort_values('year').groupby(['gwno'])['log_best'].cumsum()
    df['past_events_country'] = df.sort_values('year').groupby(['gwno'])['binary_best'].cumsum()
    
    # ----------------------------------------------------------------------------------------------
    # The loop below normalizes the features specified in the 'features_to_normalize' list
    features_to_normalize = ['gwarea', 'interp_pop_gpw_sum', 'past_fatalities_country', 'past_magnitude_country', 'past_events_country']

    for feature in features_to_normalize:
        new_name = 'norm_' + feature # Naming the normalized features
        df[new_name] = (df[feature]- df[feature].min())/(df[feature].max()-df[feature].min()) # Normalizing features        
        
    # ----------------------------------------------------------------------------------------------
    # Creating features that reflect different ways to measure past conflict relative to the size of a country.
    df['past_fatalities_country_Narea'] = df['norm_past_fatalities_country'] / (df['norm_gwarea']+1) # I add 1 in order to avoid dividing by 0
    df['past_magnitude_country_Narea'] = df['norm_past_magnitude_country'] / (df['norm_gwarea']+1)
    df['past_events_country_Narea'] = df['norm_past_events_country'] / (df['norm_gwarea']+1)
    
    # ----------------------------------------------------------------------------------------------
    # Creating a normalized feature for the size of a country's population in a given year 
    df['norm_interp_pop_gpw_sum_country'] = df.sort_values('year').groupby(['gwno'])['norm_interp_pop_gpw_sum'].transform(np.sum) 
    
#   df['past_fatalities_country_Npop'] = df['norm_past_fatalities_country'] / (df['norm_interp_pop_gpw_sum']+1) # so maybe this should be country pop not cell pop..
#   df['past_magnitude_country_Npop'] = df['norm_past_magnitude_country'] / (df['norm_interp_pop_gpw_sum']+1)
#   df['past_events_country_Npop'] = df['norm_past_events_country'] / (df['norm_interp_pop_gpw_sum']+1)

    # Creating features that reflect different ways to measure past conflict relative to the population size of a country
    df['past_fatalities_country_Npop'] = df['norm_past_fatalities_country'] / (df['norm_interp_pop_gpw_sum_country'])
    df['past_magnitude_country_Npop'] = df['norm_past_magnitude_country'] / (df['norm_interp_pop_gpw_sum_country'])
    df['past_events_country_Npop'] = df['norm_past_events_country'] / (df['norm_interp_pop_gpw_sum_country'])
    
    # ----------------------------------------------------------------------------------------------
    # Creating features with a detorriation index that reflects the effect of the conflict trap.  
    df['best_decay5'] = df.sort_values('year').groupby(['gid'])['best'].apply(decay5) # Applying the decay-function created above
    df['best_decay10'] = df.sort_values('year').groupby(['gid'])['best'].apply(decay10) # Applying the decay-function created above
    
    # Creating decay functions based on the 'log_best' feature
    df['log_best_decay5'] = df.sort_values('year').groupby(['gid'])['log_best'].apply(decay5) # Applying the decay-function created above
    df['log_best_decay10'] = df.sort_values('year').groupby(['gid'])['log_best'].apply(decay10) # Applying the decay-function created above

    # ----------------------------------------------------------------------------------------------
    # Creating a feature that measures night light emission per capita on cell lecel
    df['cell_light_Pcap'] = (df['interp_nlights_mean'])/(df['norm_interp_pop_gpw_sum']+1)
    
    
    # ----------------------------------------------------------------------------------------------
    # Creating a feature measuring the average night light emission on a country level
    df['interp_nlights_mean_country'] = df.sort_values('year').groupby(['gwno'])['interp_nlights_mean'].transform(np.sum)
#   df['norm_interp_pop_gpw_sum_country'] = df.sort_values('year').groupby(['gwno'])['norm_interp_pop_gpw_sum'].transform(np.sum)#
    
    # Creating features that measure the average night light emission relative to population and geographical area on a country level 
    df['country_light_Pcap'] = (df['interp_nlights_mean_country'])/(df['norm_interp_pop_gpw_sum_country']+1)
    df['country_light_Area'] = (df['interp_nlights_mean_country'])/(df['norm_gwarea']+1)
    
    # ----------------------------------------------------------------------------------------------
    # Creating normalized versions of earlier created features
    df['interp_nlights_mean_norm'] = (df['interp_nlights_mean'] - df['interp_nlights_mean'].min())/(df['interp_nlights_mean'].max()-df['interp_nlights_mean'].min())
    df['interp_nlights_mean_country_norm'] = (df['interp_nlights_mean_country'] - df['interp_nlights_mean_country'].min())/(df['interp_nlights_mean_country'].max()-df['interp_nlights_mean_country'].min())
    df['country_light_Pcap_norm'] = (df['country_light_Pcap'] - df['country_light_Pcap'].min())/(df['country_light_Pcap'].max()-df['country_light_Pcap'].min())
    df['country_light_Area_norm'] = (df['country_light_Area'] - df['country_light_Area'].min())/(df['country_light_Area'].max()-df['country_light_Area'].min())

    # Creating features that measures relative deprivation by  assigning a low ratio to cells with a lower n_light_mean (pcap or area) than the whole country and 1 otherwise
    df['low_ratio_light'] = np.where(df['interp_nlights_mean_norm'] < df['interp_nlights_mean_country_norm'], # if cell is below country
                                        df['interp_nlights_mean_country_norm']/df['interp_nlights_mean_norm'], # return low_ratio
                                        1) # else return 1
    
    df['low_ratio_light_Pcap'] = np.where(df['interp_nlights_mean_norm'] < df['country_light_Pcap_norm'], # if cell is below country
                                        df['country_light_Pcap_norm']/df['interp_nlights_mean_norm'], # return low_ratio
                                        1) # else return 1
    
    df['low_ratio_light_Area'] = np.where(df['interp_nlights_mean_norm'] < df['country_light_Area_norm'], # if cell is below country
                                        df['country_light_Area_norm']/df['interp_nlights_mean_norm'], # return low_ratio
                                        1) # else return 1
    
    # Replacing infinite values from low_ratio_features with the maximum low_ratio score
    features_to_remove_inf = ['low_ratio_light', 'low_ratio_light_Pcap', 'low_ratio_light_Area']
    
    #Looping over features to remove inf values
    for feature in features_to_remove_inf:
        replacement_value = df.loc[df[feature] != np.inf, feature].max() # Creating the replacement value (max low_ratio score)
        df[feature].replace(np.inf, replacement_value, inplace=True) # Replacing inf values
    
    last_year = df['year'].max()
    
    train_set = df[df['year'] < last_year]
    other_set = df[df['year'] == last_year] # val or test set

    return train_set, other_set
```

In this final part, I will introduce and justify the different features that I have created.

Overall, the new features can be grouped into two themes: (1) the role of past conflict and (2) economic conditions. 

The first group of features provides different ways to measure the role of past conflict, and can be used to understand how important previous conflict is for predicting future conflict - this is often referred to as the conflict trap (Hegre et al., 2017b). For example, Hegre et al. (2017b) points to the fact that a previous conflict event may still have an effect more than 25 years after the conflict event actually happened (Hegre et al., 2017b, 249). Hegre et al. (2017b) argue that the conflict trap works through four different paths. It increases likelihood of:
- continuation
- recurrence
- escalation
- diffusion

With this in mind, it is clear that features captuering the conflict trap are relevant to incorporate in our model if we want to predict future conflict.

The second group of features are derived from the interpolated night light emission feature that is thought to measure economic wealth. Therefore, the features reflects different ways to measure the wealth and economic activity of a country/cell. Night light emission has previously prooved to be a reasonable proxy for measuring economic development and local wealth (Widemann & Schutte, 2017). Furthormore, the economic situation of a country has also shown to have a quite robust negative relationship with the onset of civil war, therefore, it seems prudent to include it in our models (Collier & Hoeffler, 1998; Fearon & Laitin, 2003). On a site note, the use of night light emission as a proxy of economic activity also illustrates the potential of innovative solutions to dealing with missing data (Hegre et al. 2017a). 

Furthermore, some of the features are normalized versions of the features created. Therefore, I will not describe these features in as much detail, instead, the important thing to note is that the process of normalization can help to make data less sensitive to the scale of features and improve the analysis of multiple models since our data is more consistent with each other (DeFillipi, 2018). It is also prudent to normalize features when we, for example, divide them with each other as we did above. Normalization helps to make the scales somewhat similar and we avoid numerical underflow. If the numerator gets to small, the fraction will be very small and we get a value of zero, if the denominator gets to small we risk dividing by zero which will give us an error. Another very important thing, in regards to the purpose of normalizing our features, is that our machine learning models will run more efficiently when the features are normalized since the model works better with features in a similar range.
